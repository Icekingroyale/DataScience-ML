## Pandas

### Creating a DataFrame using lists

```python
import pandas as pd
import numpy as np
```

```python
my_list = [['apple','red'],['banana','yellow'],['orange','orange']]
myDataFrame = pd.DataFrame(my_list)
myDataFrame
```

```python
myDataFrame1 = pd.DataFrame(my_list, columns = ['Fruit','Color'])
print(myDataFrame1)
```

```python
mylist = np.array([[0,1],[2,3],[4,5]])
myDataFrame2 = pd.DataFrame(mylist, columns = ['even','odd'])
print(myDataFrame2)
```

### creating a DataFrame using Dictionary

```python
# using dictionaries are my prefered method of creating a data frame in pandas
myDic = {
        'fruits':['apple','orange','banana'],
        'colors':['red','blue','green'],
        'cars':['benz','toyota','ford']
        }
myDataFrame3 = pd.DataFrame(myDic)
print(myDataFrame3)
                   
```

### filtering rows

```python

#from the first 10 rows, get rows with "data_value" higher than 5000
df10 = csvfile.head(10)
condition = df10['data_value'] > 5000
df10[condition]
```

```python
#filtering using AND = & in panda
df10[(df10['data_value'] > 5000) & (df10['units'] == 'Dollars')]
```

```python
#filtering using OR = | in panda
df10[(df10['data_value'] > 5000) | (df10['magnitude'] == 'Thousands')]
```
